## **Estimating CO2 using Volume**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import r2_score

df = pd.read_excel("Regression Data.xlsx")
df.head()
```

```python
%matplotlib inline
plt.xlabel('Volume', fontsize = 15)
plt.ylabel('CO2', fontsize = 15)
plt.scatter(df.Volume, df.CO2, color = 'red')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['Volume']], df.CO2)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[1800]])
```

```python
%matplotlib inline
plt.xlabel('Volume', fontsize = 15)
plt.ylabel('CO2', fontsize = 15)
plt.scatter(df.Volume, df.CO2, color = 'red')
plt.plot(df.Volume, reg.predict(df[['Volume']]), color = 'black')
```

```python
y_test = df['CO2']
y_predicted = reg.predict(df[['Volume']])
print("R square = ", r2_score(y_test, y_predicted))
```

## Estimation of CO2 using both Volume and Weight

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score 
from sklearn.model_selection import train_test_split

df_new = df[['Weight','Volume']]
y = df[['CO2']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_new, y, test_size=0.3, random_state=45)
```

```python
reg = linear_model.LinearRegression()
reg.fit(X_train, y_train)
y_predicted = reg.predict(X_test)
plt.scatter(y_test, y_predicted, color ='black', marker = '+')
```

```python
print('R square =',r2_score(y_test, y_predicted))
```
